# Meshgrid Tutorial for beginners

#### Necessary Imports

````python
import numpy as np
import matplotlib.pyplot as plt
````

#### Specify the number of points for different dimensions that you would like to use for the computation of your formula. For example, in this tutorial, we are going to compute the formula

```
Z = X^2 + Y^2
```

To compute the above formula, we need to have some values for X and Y. If both X and Y contain only one point with the value 1 then the value of Z would simply be (1)^2 + (1)^2 = 2. This is not very helpful if we want to get the shape of the curve for Z, which is a function of X and Y. So, to generate more than one point for X and Y, we are going to use `linspace()` method provided by the `NumPy` package. To use `linspace()` method, we need to specify the minimum limit, maximum limit, and the number of points we would like to have between those limits. Let's take an example below!

**Let's generate the points for X:**

Let's consider that we would like to see the variation of `X` between `-1` to `1`. And, to see this variation, we would like to consider **11 points** between these limits (i.e. `-1` to `1`).

````python
x_values = np.linspace(start=-1, stop=1, num=11)
````

````python
# Let's see how our x_values look like:
x_values
````

The above values of `x_values` represent that our `X` is going to take above **11 values** between **-1 to 1**

**Let's generate the points for Y:**

Let's consider that we would like to see the variation of `Y` between `-3` to `3`. And, to see this variation, we would like to consider **5 points** between these limits (i.e. `-3` to `3`).

````python
y_values = np.linspace(-3, 3, 5)
````

````python
# Let's see how our x_values look like:
y_values
````

The above values of `y_values` represent that our `Y` is going to take above **5 values** between **-3 to 3**

## Let's Generate Meshgrid

**First understand the dimensions:**
Since we are interested in getting The `meshgrid`

````python
X, Y = np.meshgrid(x_values, y_values)
````

## Let's analyse Meshgrid Results:

Before explaining further, let's have a look at the matrices `X` and `Y` created by the meshgrid function.


````python
X
````

````python
Y
````

### What are `X` and `Y` representing:

1. **X**: We wanted to have 11 values for X between -1 and 1, right? This is exactly what we have in each of X. But why do we have these values **repeated 5 times** (i.e. 5 rows)? Interesting, right? We have 5 rows in `X` because we wanted to have 5 points in the y-direction (as specified in the line `y_values = np.linspace(-3, 3, 5)` above). Therefore, we have 5 rows of the same set of elements in X.

2. **Y** For `Y`, we wanted to have only 5 points between the value -3 to 3, and that's exactly what we have received in each column of the `Y`. But again, why do we have the same columns repeated 11 times? This has the same reason as described above for `X`. We have specified above in the line `x_values = np.linspace(start=-1, stop=1, num=11)` that we would like to have 11 values in the x-direction, and therefore, the `meshgrid` function has created 11 columns.

### Another kind of indexing for `X` and `Y`:

In the above scenario, the values of `x` varied alone the rows (i.e. true x-direction) in the matrix `X`, and the values in `Y` varied along the columns (i.e. true y-direction). But in certain cases, we want to flip the behavior i.e. we would like the `x` values to vary along the columns and `y` values to vary along the rows. An example of this is the scenario when we want to **traverse along the rows and columns** of an image. To access a pixel in an image, we need to pass the (row, column) coordinates, and if we want to represent `row` as `x` and `col` as `y`, we would like to get the interpretation of `X` and `Y` flipped.

To achieve that, we need to use an additional argument called `indexing="ij"` in the `meshgrid()` function.

Let's try that...


````python
X_flipped, Y_flipped = np.meshgrid(x_values, y_values, indexing='ij')
````

````python
X_flipped
````

````python
Y_flipped
````

#### ...let's analyse the **flipped** versions

**X_flipped:**
* Again we have 11 values for `x` in the ranges -1 to 1
* **BUT** Now the values in `X` are varying along the columns

**Y_flipped:**
* Again we have 5 values for `y` in the range of -3 to 3
* **BUT** the values for `y` are getting varied along the rows

## Our Ultimate Goal:

Remember that our main goal of computing these `x` and `y` values (in `X` and `Y`) was to compute a function of `x` and `y` i.e. `Z` over these values. **But** how the final values of `Z` would look like depends on the indexing scheme we have used in the `meshgrid()` function above. Let's first consider the default indexing scheme i.e. `x` values varying along x-axis and the `y` values varying along the y-axis.

````python
Z = X**2 + Y**2
Z
````

````python
plt.imshow(Z,  extent=[-3, 3, -1, 1])
````

#### ...and now for the flipped version 

````python
Z_flipped = X_flipped**2 + Y_flipped**2
Z_flipped
````

````python
plt.imshow(Z_flipped,  extent=[-3, 3, -1, 1])
````

#### ...flipped version but with **swapped** `extent`

````python
plt.imshow(Z_flipped,  extent=[-1, 1, -3, 3])
````
